```python
#> --- Imports ---

import sys
sys.path.insert(0, r"..\src\random_forest")
import pandas as pd
import datetime as dt
from random_forest import RandomForest
from random_forest import Train_Test_Split
from random_forest import Model_Evaluation
```

```python
#> --- Importing Processed Data --- <#

data_frame = pd.read_csv(r'..\data\processed\processed_transactions_dataset_v3.csv')
```

```python
#> --- Preperation for Model Train --- <#

x_array=data_frame.iloc[:,:-1].to_numpy()
y_array=data_frame.iloc[:,-1].to_numpy()

x_train, x_test, y_train, y_test = Train_Test_Split(x_array, y_array)
```

```python
#> --- Getting an Instance from the Model --- <#

RandomForestModel=RandomForest()
```

```python
#> --- Training the Model --- <#

RandomForestModel.fit(x_train, y_train, n_trees=80, trees_max_depth=50, min_leaf_purity=0.98)
```

```python
#> --- Predicting Test Array --- <#

ts1=dt.datetime.now()
predicted_y=RandomForestModel.predict(x_test)
ts2=dt.datetime.now()
print(f"Pridected test values successfully. Time Taken: {ts2-ts1}")
```

```python
#> --- Evaluating Model --- <#

Model_Evaluation(predicted_y, y_test, save=True, RFModel=RandomForestModel, text_file_name="config_1.txt") #> default_path=r"..\results\metrics"
```
